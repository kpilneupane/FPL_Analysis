```python
url = 'https://fantasy.premierleague.com/api/bootstrap-static/'
r = requests.get(url)
fpl_data = r.json()
```

```python
# fpl_data.keys()
fpl_data = fpl_data
elements_df = pd.DataFrame(fpl_data['elements'])
# elements_df.to_csv('elements.csv')
# elements_df.columns
teams_df = pd.DataFrame(fpl_data['teams'])
fpl_data

```

```python
elements_df = pd.read_csv('elements.csv', index_col=[0])
elements_df.columns
```

```python
cols = ['element_type', 'event_points', 'first_name','second_name', 'form', 'points_per_game', 'selected_by_percent', 'status', 'team',  'total_points', 'goals_scored', 'assists', 'yellow_cards', 'bps', 'now_cost', 'minutes', 'clean_sheets']
df = elements_df[cols]
df.dtypes
```

```python
cols = ['form', 'points_per_game', 'selected_by_percent']
df.loc[:, tuple(cols)].astype('float32')
```

```python
element_type_df = pd.DataFrame(fpl_data['element_types'])
pos_dict = dict(zip(element_type_df['id'], element_type_df['singular_name']))
df.loc[:, 'position'] = df.loc[:,'element_type'].apply(lambda x: pos_dict[x])

df['position'].value_counts().index

#Map team to team name
team_codes =list(df['team'].unique())
team_names = list(teams_df['short_name'].unique())
teams_dict = dict(zip(team_codes, team_names))

df.loc[:, 'G+A'] = df.loc[:, 'goals_scored'] + df.loc[:, 'assists']

df.loc[:, 'team'] = df.loc[:,'team'].apply(lambda x: teams_dict[x])
```

```python
# df.loc[:, 'price'] = df.loc[:,'now_cost']/10
# df['now_cost'].head
# df.loc[:,'now_cost']/10
df.loc[:,'price'] = df.loc[:, 'now_cost']/10
```

```python
labels = ['Budget', 'Mid', 'Premium']
#Create bins
gk_bins = [4.0, 5.0, 6.0, 7.0]
def_bins = [4.0, 5.0, 6.0, 8.0]
mid_bins = [4.0, 6.0, 9.0, 14.0]
for_bins = [4.0, 6.0, 9.0, 14.0]

tup = [(df, gk_bins, 'Goalkeeper', labels),
       (df, def_bins, 'Defender', labels),
       (df, mid_bins, 'Midfielder', labels),
       (df, for_bins, 'Forward', labels)
       ]

#Discretinization function
def categorize_price(dataframe, bins, position, labels):
    grp = dataframe.groupby(['position']).get_group(position)
    # grp_prices = dataframe.groupby(['price'].get_group(bins))
    grp_prices = grp['price']
    grp.loc[:,'price_range'] = pd.cut(grp_prices, bins, labels=labels, right=False)
    print(grp.head)
    grp_lst.append(grp)
    
    return grp

#Create grp_list
grp_lst = list()

for item in tup:
    df = item[0]
    bins = item[1]
    pos = item[2]
    labels = item[3]
    categorize_price(df, bins, pos, labels)
    
frames = grp_lst
df = pd.concat(frames)
# print(df.head)
```

```python
# creating labels
labels = ['Differential', 'High']

# Creating range
selected_by = [0.0, 15.0, 99.0]

# create list of tuples
tup = (df, selected_by, labels)

# Distinction function
def categorize_own(dataframe, bins, labels):
    ownership = dataframe['selected_by_percent']
    dataframe['ownership'] = pd.cut(ownership, bins, labels=labels, right=False)
    return dataframe

categorize_own(df, selected_by, labels)
```
